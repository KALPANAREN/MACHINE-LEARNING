```python
"""
23_apriori.py
==================================================================

APRIORI ALGORITHM

ASSOCIATION RULE MINING

Dataset:
Synthetic Retail Transactions

REAL WORLD WORKFLOW

1. Data Understanding
2. EDA
3. Transaction Encoding
4. Frequent Itemset Mining
5. Support Analysis
6. Apriori Model Building
7. Rule Generation
8. Confidence Analysis
9. Lift Analysis
10. Rule Filtering
11. Business Interpretation
12. Recommendation Examples
13. Rule Stability
14. Deployment Readiness
15. Interview Questions

==================================================================

Required Package:

pip install mlxtend

==================================================================
"""

# ==========================================================
# STEP 0 : IMPORTS
# ==========================================================

import pandas as pd
import numpy as np

from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import (
    apriori,
    association_rules
)

# ==========================================================
# STEP 1 : DATA UNDERSTANDING
# ==========================================================

"""
Each row represents

ONE CUSTOMER TRANSACTION
"""

transactions = [

    ['Milk','Bread','Butter'],
    ['Bread','Butter'],
    ['Milk','Bread'],
    ['Milk','Butter'],
    ['Bread','Eggs'],
    ['Milk','Bread','Butter','Eggs'],
    ['Milk','Bread','Eggs'],
    ['Bread','Butter','Eggs'],
    ['Milk','Bread','Butter'],
    ['Milk','Eggs'],
    ['Bread','Butter'],
    ['Milk','Bread','Butter','Eggs'],
    ['Milk','Bread'],
    ['Butter','Eggs'],
    ['Milk','Bread','Butter'],
    ['Milk','Bread','Butter','Eggs'],
    ['Milk','Eggs'],
    ['Bread','Eggs'],
    ['Milk','Bread','Butter'],
    ['Milk','Bread']
]

print("Total Transactions")

print(len(transactions))

# ==========================================================
# STEP 2 : EDA
# ==========================================================

all_items = []

for transaction in transactions:

    all_items.extend(transaction)

item_frequency = pd.Series(
    all_items
).value_counts()

print("\nItem Frequency")

print(item_frequency)

# ==========================================================
# STEP 3 : TRANSACTION ENCODING
# ==========================================================

"""
Apriori requires

One Hot Encoding
"""

te = TransactionEncoder()

te_array = te.fit(
    transactions
).transform(
    transactions
)

df = pd.DataFrame(
    te_array,
    columns=te.columns_
)

print("\nEncoded Dataset")

print(df.head())

# ==========================================================
# STEP 4 : FREQUENT ITEMSET MINING
# ==========================================================

"""
Support

=

(Itemset Occurrence)

/

(Total Transactions)
"""

frequent_itemsets = apriori(

    df,

    min_support=0.2,

    use_colnames=True

)

print("\nFrequent Itemsets")

print(frequent_itemsets)

# ==========================================================
# STEP 5 : SUPPORT ANALYSIS
# ==========================================================

frequent_itemsets = frequent_itemsets.sort_values(

    by="support",

    ascending=False

)

print("\nSorted By Support")

print(frequent_itemsets)

# ==========================================================
# STEP 6 : APRIORI MODEL BUILDING
# ==========================================================

"""
Already performed through:

apriori()

Main Hyperparameter:

min_support
"""

# ==========================================================
# STEP 7 : RULE GENERATION
# ==========================================================

rules = association_rules(

    frequent_itemsets,

    metric="confidence",

    min_threshold=0.5

)

print("\nAssociation Rules")

print(rules)

# ==========================================================
# STEP 8 : CONFIDENCE ANALYSIS
# ==========================================================

"""
Confidence

P(B|A)

Probability customer buys B
given A was purchased
"""

rules = rules.sort_values(

    by="confidence",

    ascending=False

)

print("\nRules Sorted By Confidence")

print(

    rules[[

        "antecedents",

        "consequents",

        "support",

        "confidence",

        "lift"

    ]]

)

# ==========================================================
# STEP 9 : LIFT ANALYSIS
# ==========================================================

"""
MOST IMPORTANT BUSINESS METRIC

Lift > 1

Positive Association

Lift = 1

Independent

Lift < 1

Negative Association
"""

rules = rules.sort_values(

    by="lift",

    ascending=False

)

print("\nRules Sorted By Lift")

print(

    rules[[

        "antecedents",

        "consequents",

        "support",

        "confidence",

        "lift"

    ]]

)

# ==========================================================
# STEP 10 : RULE FILTERING
# ==========================================================

strong_rules = rules[

    (rules["confidence"] >= 0.7)

    &

    (rules["lift"] >= 1.1)

]

print("\nStrong Rules")

print(

    strong_rules[[

        "antecedents",

        "consequents",

        "support",

        "confidence",

        "lift"

    ]]

)

# ==========================================================
# STEP 11 : BUSINESS INTERPRETATION
# ==========================================================

print("\nBusiness Interpretation")

for _, row in strong_rules.iterrows():

    print()

    print(

        f"If customer buys "

        f"{set(row['antecedents'])}"

    )

    print(

        f"Recommend "

        f"{set(row['consequents'])}"

    )

    print(

        f"Confidence: "

        f"{row['confidence']:.2f}"

    )

    print(

        f"Lift: "

        f"{row['lift']:.2f}"

    )

# ==========================================================
# STEP 12 : RECOMMENDATION EXAMPLES
# ==========================================================

customer_basket = {'Milk','Bread'}

print("\nCustomer Basket")

print(customer_basket)

recommendations = []

for _, row in strong_rules.iterrows():

    antecedent = set(
        row["antecedents"]
    )

    consequent = set(
        row["consequents"]
    )

    if antecedent.issubset(
        customer_basket
    ):

        recommendations.extend(
            list(consequent)
        )

print("\nRecommendations")

print(
    list(
        set(recommendations)
    )
)

# ==========================================================
# STEP 13 : RULE STABILITY
# ==========================================================

"""
Experiment

Different support thresholds
"""

supports = [

    0.1,
    0.2,
    0.3,
    0.4
]

for support in supports:

    itemsets = apriori(

        df,

        min_support=support,

        use_colnames=True

    )

    print(

        f"\nSupport={support}"

    )

    print(

        f"Frequent Itemsets={len(itemsets)}"

    )

# ==========================================================
# STEP 14 : DEPLOYMENT READINESS
# ==========================================================

"""
Production Pipeline

Customer Basket
       ↓
Association Rules
       ↓
Recommended Products

Applications

Amazon

Flipkart

BigBasket

Instacart

Walmart
"""

strong_rules.to_csv(

    "association_rules.csv",

    index=False

)

print("\nSaved")

print("association_rules.csv")

# ==========================================================
# STEP 15 : INTERVIEW QUESTIONS
# ==========================================================

"""
1. What is Apriori?

2. What is Association Rule Mining?

3. What is Support?

4. Formula for Support?

5. What is Confidence?

6. Formula for Confidence?

7. What is Lift?

8. Formula for Lift?

9. Why Lift is important?

10. Difference between
    Support and Confidence?

11. What is Frequent Itemset?

12. What is Candidate Generation?

13. What is Apriori Property?

14. What is Downward Closure Property?

15. Why Apriori is expensive?

16. What is Market Basket Analysis?

17. Real-world use cases?

18. Amazon recommendation example?

19. Lift > 1 meaning?

20. Lift < 1 meaning?

21. Confidence = 1 meaning?

22. How do you tune min_support?

23. Difference between Apriori and FP-Growth?

24. What are antecedents?

25. What are consequents?

26. Explain Apriori end-to-end.
"""

# ==========================================================
# INTERVIEW CHEAT SHEET
# ==========================================================

"""
SUPPORT

Frequency of itemset

------------------------------------------------

CONFIDENCE

Probability of buying B
given A purchased

------------------------------------------------

LIFT

How much more likely
B occurs with A

------------------------------------------------

Example

Milk -> Bread

Support = 0.40

Confidence = 0.80

Lift = 1.25

Interpretation:

Customers buying Milk
are 1.25x more likely
to buy Bread.

------------------------------------------------

Most Important Metric:

LIFT

because it removes
popularity bias.
"""

# ==========================================================
# END OF FILE
# ==========================================================
```
