```python
"""
20_pca.py
==================================================================

PRINCIPAL COMPONENT ANALYSIS (PCA)

Dataset:
Wine Dataset

REAL WORLD WORKFLOW

1. Data Understanding
2. EDA
3. Feature Selection
4. Scaling
5. Baseline PCA
6. Explained Variance Analysis
7. Component Selection
8. Cumulative Variance Analysis
9. Scree Plot
10. PCA Transformation
11. Feature Contribution Analysis
12. Visualization
13. Compare Before vs After PCA
14. Deployment Readiness
15. Interview Questions

==================================================================
"""

# ==========================================================
# STEP 0 : IMPORTS
# ==========================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA

import joblib

# ==========================================================
# STEP 1 : DATA UNDERSTANDING
# ==========================================================

wine = load_wine()

df = pd.DataFrame(
    wine.data,
    columns=wine.feature_names
)

print("\nHEAD")
print(df.head())

print("\nINFO")
print(df.info())

print("\nDESCRIBE")
print(df.describe())

print("\nSHAPE")
print(df.shape)

# ==========================================================
# STEP 2 : EDA
# ==========================================================

print("\nMISSING VALUES")
print(df.isnull().sum())

print("\nDUPLICATES")
print(df.duplicated().sum())

plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Matrix")

plt.show()

"""
Notice:

Many features are highly correlated.

PCA helps remove redundancy.
"""

# ==========================================================
# STEP 3 : FEATURE SELECTION
# ==========================================================

X = df.copy()

# ==========================================================
# STEP 4 : SCALING
# ==========================================================

"""
MOST IMPORTANT STEP

PCA uses variance.

Features with larger scale dominate.

Therefore scaling is mandatory.
"""

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# ==========================================================
# STEP 5 : BASELINE PCA
# ==========================================================

"""
First fit PCA with all features
to understand variance distribution.
"""

pca_full = PCA()

pca_full.fit(X_scaled)

# ==========================================================
# STEP 6 : EXPLAINED VARIANCE ANALYSIS
# ==========================================================

explained_variance = (
    pca_full.explained_variance_ratio_
)

print("\nExplained Variance Ratio")

for i, var in enumerate(
    explained_variance,
    start=1
):

    print(
        f"PC{i}: {var:.4f}"
    )

# ==========================================================
# STEP 7 : COMPONENT SELECTION
# ==========================================================

"""
Common Industry Rules

80% Variance
90% Variance
95% Variance

Most common:
95%
"""

cumulative_variance = np.cumsum(
    explained_variance
)

n_components_95 = np.argmax(
    cumulative_variance >= 0.95
) + 1

print(
    "\nComponents needed for 95% variance:",
    n_components_95
)

# ==========================================================
# STEP 8 : CUMULATIVE VARIANCE ANALYSIS
# ==========================================================

variance_df = pd.DataFrame({

    "Component":
    range(
        1,
        len(explained_variance)+1
    ),

    "Variance":
    explained_variance,

    "Cumulative":
    cumulative_variance

})

print("\nVariance Table")

print(variance_df)

# ==========================================================
# STEP 9 : SCREE PLOT
# ==========================================================

"""
MOST ASKED PCA INTERVIEW QUESTION

Scree Plot helps identify
optimal number of components.
"""

plt.figure(figsize=(10,5))

plt.plot(

    range(
        1,
        len(explained_variance)+1
    ),

    explained_variance,

    marker='o'

)

plt.xlabel("Principal Components")

plt.ylabel("Explained Variance")

plt.title("Scree Plot")

plt.show()

# ==========================================================
# STEP 10 : PCA TRANSFORMATION
# ==========================================================

pca = PCA(
    n_components=n_components_95
)

X_pca = pca.fit_transform(
    X_scaled
)

print("\nOriginal Shape")

print(X_scaled.shape)

print("\nPCA Shape")

print(X_pca.shape)

print(
    "\nDimensionality Reduced"
)

```

```python
# ==========================================================
# STEP 11 : FEATURE CONTRIBUTION ANALYSIS
# ==========================================================

"""
Feature loadings tell us

which features contribute
most to each component.
"""

loadings = pd.DataFrame(pca.components_.T,
    columns=[
        f"PC{i+1}"
        for i in range(n_components_95)],index=df.columns)
print("\nFeature Loadings")
print(loadings)

# Top Contributors

for pc in loadings.columns:
    print(f"\nTop Features for {pc}")
    print(loadings[pc].abs().sort_values(ascending=False).head(5))
```

```python
# ==========================================================
# STEP 12 : VISUALIZATION
# ==========================================================

"""
Most common PCA usage
Visualization in 2D
"""

pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(10,6))
plt.scatter(X_2d[:,0],X_2d[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("2D PCA Projection")
plt.show()
```

```python
# ==========================================================
# STEP 13 : BEFORE VS AFTER PCA
# ==========================================================

print("\nCOMPARISON")
print("Original Features:",X.shape[1])
print("PCA Features:",X_pca.shape[1])
print("Variance Retained:",round(cumulative_variance[n_components_95-1] * 100,2),"%")
```

```python
# ==========================================================
# STEP 14 : DEPLOYMENT READINESS
# ==========================================================

joblib.dump(scaler,"pca_scaler.pkl")
joblib.dump(pca,"pca_model.pkl")
print("\nArtifacts Saved")
print("pca_scaler.pkl")
print("pca_model.pkl")

# Example New Data

sample = X.iloc[[0]]
sample_scaled = scaler.transform(sample)
sample_pca = pca.transform(sample_scaled)

print("\nPCA Representation")
print(sample_pca)
```
